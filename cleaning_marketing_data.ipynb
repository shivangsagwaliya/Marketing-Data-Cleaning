```python
import pandas as pd
import numpy as np

## Loading the messy data
df=pd.read_csv('marketing_campaign_data_messy.csv')
print(f'Loading Dataset:{df.shape[0]} rows and {df.shape[1]} columns')
```

```python
df
```

## Step 1. Header Cleaninng

```python
df.columns=df.columns.str.strip().str.lower().str.replace(' ','_')
df.columns
```

## Step 2. Type conversion and currency cleaning

```python
dirty_spend_mask=df['spend'].astype(str).str.contains('\$')
df.loc[dirty_spend_mask,['campaign_id','spend']].head()
```

```python
df.loc[dirty_spend_mask,'spend']=df.loc[dirty_spend_mask,'spend'].str.replace(r'[^\d.-]','',regex=True)
print(f'Cleaned spend column')
```

```python
dirty_spend_mask=df['spend'].astype(str).str.contains('\$')
df.loc[dirty_spend_mask,['campaign_id','spend']].head()
```

```python
df
```

## Step 3: Categorical Typos (Fuzzy Logic)

```python
df['channel'].unique()
```

```python
df['channel']=df.channel.str.replace('_','')
```

```python
df['channel'].unique()
```

```python
df['channel']=df['channel'].apply(lambda x : 'Facebook' if x=='Facebok' else('Google' if x == 'Gogle'else x))
```

```python
df['channel'].unique()
```

```python
cleanup_mask={'Facebok':'Facebook',
              'Gogle':'Google',
              'E-mail':'Email',
              'Tik_Tok':'TikTok',
              'Insta_gram':'Instagram',
              'nan':np.nan}
df['channel']=df['channel'].replace(cleanup_mask)
print(f'Fix Applied Cleaned channel column:{df.channel.unique()}')
```

```python
df
```

## Step 4. Handling Mixed Booleans

```python
df.active.unique()
```

```python
bool_map={'Yes':True,
          'Y':True,
          '1':True,
          'True':True,
          'False':False,
          'No':False,
          'N':False,
          '0':False}
df['active']=df['active'].replace(bool_map)
print(f'Fix Applied Cleaned active column:{df.active.unique()}')
```

## Step 5: Date Parsing

```python
print(df['start_date'].dtype)
print(df['end_date'].dtype)

df['start_date']=pd.to_datetime(df['start_date'],errors='coerce')
df['end_date']=pd.to_datetime(df['end_date'],errors='coerce')

print(f"Date parsing fixed for start_date{df['start_date'].dtype} and end_date {df['end_date'].dtype}")
```

## Step 6: Logical Integrity ( Clicks vs Impressions)

```python
df=df.loc[:,~df.columns.duplicated()]
```

```python
impossible_clicks_mask=df['impressions']<df['clicks']
df.loc[impossible_clicks_mask,['impressions','clicks']]
```

## Step: 7  Logical Integrity (Time Travel)**bold text**

```python
time_travel_mask=df['start_date']>df['end_date']
df.loc[time_travel_mask,['campaign_id','start_date','end_date']].head()
```

```python
df.loc[time_travel_mask,['end_date']]=df.loc[time_travel_mask,['start_date']]+pd.Timedelta(days=30)
```

## Step 8: Handling Outliers

```python
df['spent']=df['spend'].astype(float)
df.drop(columns='spent')
```

```python
df['spend'] = df['spend'].astype(float)
Q1= df['spend'].quantile(0.25)
Q3= df['spend'].quantile(0.75)
IQR = Q3 - Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

outlier_mask=df['spend']>upper_bound
df.loc[outlier_mask,['spend']]
```

```python
df[outlier_mask,'spend']=upper_bound
```
